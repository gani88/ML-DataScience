## Dictionary
Just for refresher

Why Do We Need Dictionaries?

Dictionaries in Python are essential data structures that provide a way to store and retrieve data efficiently. They are used for several reasons:

- Key-Value Mapping: Dictionaries allow to associate a key with a value. This association is often referred to as a "key-value pair." Keys are unique within a dictionary, and each key is associated with a specific value. This key-value mapping is useful for organizing and retrieving data based on meaningful labels.

- Faster Data Retrieval: Dictionaries are highly efficient for data retrieval. Unlike lists or arrays, where when access to elements by their position (index), dictionaries allow to access values directly by their keys. This means that regardless of the dictionary's size, looking up a value by its key takes a constant amount of time, making dictionaries well-suited for tasks like indexing and searching.

````python
# Creating a dictionary
dictionary = {'spain': 'madrid', 'france': 'paris', 'italy': 'rome'}

# Accessing values using keys
capital_of_spain = dictionary['spain']  # Retrieves the value associated with the key 'spain'

# Printing the result
print(f"Key is 'spain'. Value is {capital_of_spain}.")
````

- The dictionary contains three key-value pairs: 'spain': 'madrid', 'france': 'paris', and 'italy': 'rome'.
- When we use the key 'spain' to access the dictionary, we get the corresponding value 'madrid'.
- The key 'spain' is a label or identifier for the value 'madrid'.

````python
# Keys must be immutable objects like string, boolean, float, int
# List is not immutable
# Key must unique

dictionary['spain'] = "barcelona"
print(dictionary)

dictionary['usa'] = "vegas"
print(dictionary)

del dictionary['spain']
print(dictionary)

print('france' in dictionary)

dictionary.clear()
print(dictionary)
````

## Scope

Scope in Python refers to where code a particular variable is recognized and can be used. There are three main types of scope to be aware of:

- Global Scope: Variables defined in the main body of a script, outside of any function, have a global scope. Meaning they can be accessed from anywhere in code, both inside and outside functions.


```
x = 10  # This variable has global scope

def my_function():
    print(x)  # You can access x here because it's in global scope

```

- Local Scope: Variables defined within a function have a local scope. These variables are only accessible within that specific function.


```
def my_function():
    y = 5  # This variable has local scope
    print(y)  # You can access y within this function, but not outside

my_function()
# print(y)  # This would result in an error because y is not defined in this scope

```

- Built-in Scope: Python provides a set of built-in functions and modules that are always available, such as print() and len(). These functions have a built-in scope, and you can use them without defining them.


```
print("Hello, world!")  # Using the built-in print function
length = len("Python")  # Using the built-in len function

```

In simple terms, global scope is like having a variable available everywhere in your program, local scope restricts a variable's availability to just within a specific function, and built-in scope includes functions and modules that are always ready to use without needing to define them.

Variables with the same name in different scopes, the one in the local scope takes precedence. If using the global variable with the same name inside a function, may need to explicitly indicate that it's a global variable using the global keyword.

````python
# Example 1

x = 2
def f():
  x = 3
  return x

print(x)   # global scope
print(f()) # local scope
````

````python
# Example 2

x = 5
def f():
  y = 2*x
  return y

print(f()) # Will use global scope x
````

## Default & Flexible Arguments

**Default Argument**

When define a function in Python, we can give some of its parameters default values. These default values are used when we don't provide a specific value for that parameter when calling the function. Here's an example:

```
def greet(name, greeting="Hello"):
    """'greeting' has a default value of 'Hello'."""
    print(greeting, name)

greet("Alice")  # Output: Hello Alice
greet("Bob", "Hi")  # Output: Hi Bob

```
In this example, the greeting parameter has a default value of "Hello." If don't provide a value for greeting, it uses the default value.

---

**Flexible Argument (Using *args)**

Sometimes, we create a function that can accept a variable number of arguments. We can use *args to achieve this. It allows  to pass one or more arguments to the function. Here's an example:

```
def add(*args):
    """*args can take one or more numbers and add them."""
    result = 0
    for num in args:
        result += num
    return result

sum1 = add(1, 2)  # Sum of 1 and 2 = 3
sum2 = add(1, 2, 3, 4, 5)  # Sum of 1, 2, 3, 4, and 5 = 15

```
In this case, *args allows  to pass as many numbers as we want to the add function, and it will add them together.

---
**Keyword Argument  (Using **kwargs)**

When pass a variable number of keyword arguments to a function, we can use **kwargs. It turns the keyword arguments into a dictionary. Here's an example:


```
def person_info(**kwargs):
    """**kwargs stores information about a person."""
    for key, value in kwargs.items():
        print(key, ":", value)

person_info(name="Alice", age=30, city="New York")

```

In this example, **kwargs allows  to pass key-value pairs as arguments to the person_info function. It treats them as a dictionary, so it can access the information within the function.


In summary, default arguments provide default values for parameters in a function. *args allows a function to accept a variable number of positional arguments, and **kwargs allows a function to accept a variable number of keyword arguments, treating them as a dictionary. These concepts make functions more flexible and versatile in Python.

````python
# Example default parameter

def f(a, b=1, c=2):
  y = a + b + c

  return y

print(f(15))

# trying to change the default
print(f(6,2,3))
````

````python
# Example *args

def f(*args):
  for i in args:
    print(i)

f(1)
print("")
f(1,2,3,4)
````

````python
# Example **kwargs

def f(**kwargs):
  """
  print key and value dict
  """

  for key, value in kwargs.items():
    print(key, " ", value)

f(country='spain', capital='madrid', pop=782341)
````

## Anon Function

More arguments.
Ex :

map(func,seq) : applies a function to all the items in a list.

The map function is used to apply a function to every item in an iterable, such as a list, and return the results as a new iterable. It takes two arguments:

**the function you want to apply and the iterable you want to apply it to**



```
# Syntax: map(function, iterable)
numbers = [1, 2, 3, 4, 5]
squared = map(lambda x: x**2, numbers)
# 'squared' is now an iterable containing the squares of 'numbers'

# To see the results, you can convert 'squared' to a list
squared_list = list(squared)
# 'squared_list' contains [1, 4, 9, 16, 25]

```
In this example, we used the map function to apply a lambda function to each element in the numbers list, squaring each number. The result is an iterable, which we converted into a list to see the squared values.

The map function is a convenient way to apply a function to all items in a sequence and collect the results without having to write explicit loops.


---


In summary, anonymous functions (lambda functions) in Python are small, nameless functions that can take multiple arguments. The map function is used to apply a function to all items in an iterable, creating a new iterable containing the results of applying the function to each item.

## List Comp

- List comprehension is a concise and powerful way to create lists in Python.
- It allows  to generate a new list by applying an expression to each item in an existing iterable (e.g., a list, string, or range) in a single line of code.
- It essentially collapses a for loop used to build a list into a single line.

In next example, we create num2 by adding 1 to each element in num1 using list comprehension. It's a more concise and readable way to generate a new list. List comprehensions are commonly used in data analysis and are great for creating new lists by transforming or filtering existing data.

````python
# Using traditional loop
num1 = [1,2,3]
num2 = []

for n in num1:
  num2.append(n+1)
num2
````

````python
# Using List Comp

num1 = [1,2,3]
num2 = [i+1 for i in num1]

print(num2)
````

- [i + 1 for i in num1 ]: list of comprehension
- i +1: list comprehension syntax
- for i in num1: for loop syntax
- i: iterator
- num1: iterable object

````python
# Conditional on iterable

num1 = [5,10,15]
num2 = [i**2 if i == 10 else i-5 if i < 7 else i+5 for i in num1]
print(num2)
````

## EDA (Exploratory Data Analysis)

Terms that important :

- value_counts(): It's a method to count how often each unique value appears in a dataset, which is useful for understanding the distribution of data.

- Outliers: These are data points that are significantly higher or lower than the rest of the data. They can skew analysis process.

- Identifying Outliers: To find outliers, we can use the Interquartile Range (IQR) method. To calculate the IQR as Q3 (the value at 75%) minus Q1 (the value at 25%). Outliers are values smaller than Q1 - 1.5 times IQR or larger than Q3 + 1.5 times IQR.

- describe() Method: This method provides a quick summary of the data, including the count of entries, the mean (average), the standard deviation, the minimum and maximum values, and three quartiles (Q1, Q2, Q3).

In summary, value_counts() helps to count unique values, outliers are extreme values in  data, and we can use the describe() method to get a quick overview of the dataset's statistics, including quartiles for outlier detection.
