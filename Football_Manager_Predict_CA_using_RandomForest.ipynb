```python
# Load Library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
# Load data
att_data_path = "/content/variables.csv"
att_data = pd.read_csv(att_data_path)
```

```python
# Cleaning Data
att_data = att_data.dropna(axis=0)
```

```python
att_data.describe()
```

```python
att_data.columns
```

```python
# Target & Features
y = att_data.CA
att_data_features = ['Apps', 'Height', 'Weight', 'Age', 'Av Rat', 'Gls', 'xG']
X = att_data[att_data_features]
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
```

```python
train_X
```

```python
val_X
```

```python
forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
stat_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, stat_preds))
```
